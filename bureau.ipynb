```python
import pandas as pd
```

```python
df = pd.read_csv('data/application_train.csv')
bureau = pd.read_csv('data/bureau.csv')
bureau['id'] = range(len(bureau))
bureau = pd.get_dummies(bureau)
counts = df[['SK_ID_CURR']].merge(bureau, how='left', on='SK_ID_CURR').groupby('SK_ID_CURR').agg(
    {'id': 'count', 'CREDIT_ACTIVE_Active': 'sum', 'AMT_CREDIT_SUM': 'mean'})
```
